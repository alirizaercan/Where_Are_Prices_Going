```python
df = pd.read_csv('akaryakit_fiyatlari_extracted.csv', parse_dates=['Tarih'], dayfirst=True)

# 'Kurşunsuz Benzin (TL/LT)' sütununu sayısal formata çevirin
df['Kurşunsuz Benzin (TL/LT)'] = pd.to_numeric(df['Kurşunsuz Benzin (TL/LT)'].str.replace(',', '.'))

# Aylık gruplama ve ortalamaları alma
df_monthly = df.resample('M', on='Tarih').mean()

# Zaman serisi grafiği oluşturun
plt.figure(figsize=(12, 6))
plt.plot(df_monthly.index, df_monthly['Kurşunsuz Benzin (TL/LT)'], marker='o', color='blue')
plt.title('Kurşunsuz Benzin Fiyatları Zaman Serisi (Aylık Ortalamalar)')
plt.xlabel('Tarih')
plt.ylabel('Fiyat (TL/LT)')
plt.grid(True)
plt.show()
```

```python
# Verilerinizi içeren CSV dosyasını okuyun
df = pd.read_csv('akaryakit_fiyatlari_extracted.csv', parse_dates=['Tarih'], dayfirst=True)

# 'Kurşunsuz Benzin (TL/LT)' sütununu sayısal formata çevirin
df['Kurşunsuz Benzin (TL/LT)'] = pd.to_numeric(df['Kurşunsuz Benzin (TL/LT)'].str.replace(',', '.'))

# Aylık gruplama ve ortalamaları alma
df_monthly = df.resample('M', on='Tarih').mean()

# Zaman serisi grafiği oluşturun
plt.figure(figsize=(12, 6))
plt.plot(df_monthly.index, df_monthly['Kurşunsuz Benzin (TL/LT)'], color='blue', linestyle='-')
plt.title('Kurşunsuz Benzin Fiyatları Zaman Serisi (Aylık Ortalamalar)')
plt.xlabel('Tarih')
plt.ylabel('Fiyat (TL/LT)')
plt.grid(True)
plt.savefig('ay_ve_yillara_gore_akaryakit_fiyati.png')
plt.show()
```

# Asgari Ucrete Gore Veri Gorsellestirilmesi

```python
# Veriyi yükle
df = pd.read_csv('asgari_ucret_adjusted.csv')

# Tarih sütununu datetime formatına çevir
df['Start Date'] = pd.to_datetime(df['Start Date'])

# Çizgi grafiği oluştur
plt.figure(figsize=(10, 6))
plt.plot(df['Start Date'], df['16 Yaşını Dolduranlar için Net Asgari Ücret'], marker='o', linestyle='-', color='b')

# Grafiği özelleştir
plt.title('Net Asgari Ücret - Zaman İçinde Değişim')
plt.xlabel('Tarih')
plt.ylabel('Net Asgari Ücret')
plt.grid(True)

# Grafiği göster
plt.show()
```

```python
# Verilerinizi içeren CSV dosyasını okuyun
df = pd.read_csv('asgari_ucret_adjusted.csv', parse_dates=['Start Date'], dayfirst=True)

# 'Start Date' sütununu tarih tipine çevirin
df['Start Date'] = pd.to_datetime(df['Start Date'])

# '16 Yaşını Dolduranlar için Net Asgari Ücret' sütununu sayısal formata çevirin
df['16 Yaşını Dolduranlar için Net Asgari Ücret'] = pd.to_numeric(df['16 Yaşını Dolduranlar için Net Asgari Ücret'].str.replace('.', '').str.replace(',', '.'))

# Zaman serisi grafiği oluşturun
plt.figure(figsize=(12, 6))
plt.plot(df['Start Date'], df['16 Yaşını Dolduranlar için Net Asgari Ücret'], marker='o', linestyle='-', color='b')
plt.title('Net Asgari Ücret Zaman Serisi')
plt.xlabel('Tarih')
plt.ylabel('Net Asgari Ücret')
plt.grid(True)
plt.show()
```

```python
# Verilerinizi içeren CSV dosyasını okuyun
df = pd.read_csv('asgari_ucret_adjusted.csv', parse_dates=['Start Date'], dayfirst=True)

# 'Start Date' sütununu tarih tipine çevirin
df['Start Date'] = pd.to_datetime(df['Start Date'])

# '16 Yaşını Dolduranlar için Net Asgari Ücret' sütununu sayısal formata çevirin
df['16 Yaşını Dolduranlar için Net Asgari Ücret'] = pd.to_numeric(df['16 Yaşını Dolduranlar için Net Asgari Ücret'].str.replace('.', '').str.replace(',', '.'))

# 2010-01-01 tarihinden sonraki verileri filtreleyin
df_filtered = df[df['Start Date'] >= '2010-01-01']

# Zaman serisi grafiği oluşturun
plt.figure(figsize=(12, 6))
plt.plot(df_filtered['Start Date'], df_filtered['16 Yaşını Dolduranlar için Net Asgari Ücret'], linestyle='-', color='b')
plt.title('Net Asgari Ücret Zaman Serisi (2010-01-01\'den Sonra)')
plt.xlabel('Tarih')
plt.ylabel('Net Asgari Ücret')
plt.grid(True)
plt.savefig('ay_ve_yillara_gore_asgari_ucret.png')
plt.show()
```
